```python
# 第二大题
import pandas as pd
import numpy as np

# 第一题
iris = pd.read_csv("iris.csv")
iris.columns = ['sepal_length','sepal_width','petal_length','petal_width','calss']
# iris

# 第二题
iris.loc[0:9,'petal_length'] = np.nan
# iris.head(20)

# 第三题
iris.fillna(1.0,inplace=True)
# iris.head(15)

# 第四题
iris.drop(columns='calss',inplace=True)
# iris

# 第五题
iris.loc[0:2,:] = np.nan
# iris

# 第六题
iris.dropna(inplace=True)
# iris


# 第七题
iris.reset_index(inplace=True,drop=True)
# iris

# 第七题
iris.to_csv('iris_new.csv')
```

```python
# 第三大题

import pandas as pd
import matplotlib.pyplot as plt

# 第一题
wine = pd.read_csv('wine.csv')
# wine

# 第二题
wine.isnull().sum().sum()

# 第三题
value = np.array(wine)
data = [wine.loc[:,"Class"].value_counts()[1],wine.loc[:,"Class"].value_counts()[2],wine.loc[:,"Class"].value_counts()[3]]
labels = ['1','2','3']
plt.figure(figsize=(8,5))
plt.pie(data,labels=labels,explode=[0.03,0.03,0.03],autopct="%1.2f%%")
plt.show()

# 第四题
wine_data = wine.iloc[:,1:]
wine_label = wine.iloc[:,0]

# 
from sklearn.model_selection import train_test_split

wine_train,wine_test,wine_train_label,wine_test_label=train_test_split(wine_data,wine_label,test_size=0.2)

# 
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3)
kmeans.fit(wine_train)

# 
from sklearn.metrics import fowlkes_mallows_score

fowlkes_mallows_score(wine_train_label,kmeans.labels_)

# 
for i in range(2,11):
    kmeans=KMeans(n_clusters=i).fit(wine_train)
    print(fowlkes_mallows_score(wine_train_label,kmeans.labels_))
```
