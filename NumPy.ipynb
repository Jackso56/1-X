#### numpy.arange()

```python
# arange([start,] stop[, step,], dtype=None, *, like=None)
# step=10
np.arange(10)
```

```python
# start=1   stop=10
np.arange(1,10)
```

#### numpy.linspace()

```python
# linspace(start, stop, num=50, endpoint=True, 
#          retstep=False, dtype=None, axis=0)
# 生成等差数组   num表示产生数据的个数
```

```python
# 默认产生50个数据
np.linspace(1,10)
```

```python
# 指定元素产生个数
np.linspace(1,10,4)
```

#### numpy.logspace()

```python
# logspace(start, stop, num=50, endpoint=True, 
#           base=10.0, dtype=None, axis=0)
# num表示数据的数量， base表示公比
```

```python
# 公比默认为10，默认产生50个元素
# start、stop 表示指数的起止
np.logspace(1,10,num=4,base=2)
# [   2.,   16.,  128., 1024.]
#    2^1    2^4   2^7   2^10
```

#### 案例

```python
n = 100
data = [i for i in range(100)]
x = data
y = data
result = np.zeros([n,n])
# x,y,result
for i in range(n):
    for j in range(n):
        result[i,j] = np.sqrt(
                (x[i]-x[j])**2+
                (y[i]-y[j])**2)
result
```

#### 数组分割

### 通用函数的广播机制

```python
# 广播机制
arr1 = np.array([number for number in range(12)]).reshape(3,4)
arr2 = np.array([1,2,3,4])
arr1,arr2
```

```python
# 一维数组的广播机制：按行依次相加
arr1+arr2
```

```python
# 广播机制
arr3 = np.array([number for number in range(3)]).reshape(3,1)
arr4 = np.array([number for number in range(12)]).reshape(3,4)
arr3,arr4
```

```python
# 二维数组的广播机制：按列依次相加
arr3+arr4
```

## numpy实现文件的读取

### 二进制文件的读取

```python
import numpy as np
```

```python
arr = np.arange(10)
arr
```

#### numpy.save()

```python
# 数据保存  单数组的保存
np.save('arr_save.npy',arr)
```

#### numpy.savez()

```python
arr1 = np.random.randint(1,8,10)
arr2 = np.random.randint(1,9,10)
# 数据保存 多数组保存
np.savez('arr_savez.npz',arr1,arr2)
```

#### numpy.load()

```python
np.load('arr_save.npy')
```

```python
# files属性查看数组的索引
np.load('arr_savez.npz').files
```

```python
# 索引方式读取数据
np.load('arr_savez.npz')['arr_1']
```
