# Matplotlib模块

## 添加子图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.arange(1,100)
data
```

```python
plt.rcParams['font.sans-serif']='SimHei'
p=plt.figure(figsize=(8,4))
# add_subplot(行数，列数，第几幅图)
p.add_subplot(1,2,1)
plt.plot(data,1/data,c='c',ls='-.',alpha=0.4)
plt.xlabel('x')
plt.ylabel('1/x')
plt.title("1/x")
plt.legend(['y=1/x'])
p.add_subplot(1,2,2)
plt.plot(data,data**2,c='m',ls='-.',alpha=0.3)
plt.xlabel('x')
plt.ylabel('x^2')
plt.title("x^2")
plt.legend(['y=x^2'])
plt.show()
```

## 添加文本、变换绘图主题

```python
data
```

```python
# 查看绘图样式
print(plt.style.available)
```

```python
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(5,3))
# 变换主题样式
plt.style.use(plt.style.available[10])
plt.title("添加文本、变换主题演示")
plt.plot(data,3/data,c='b',ls='--',alpha=0.4)
plt.xlabel('x')
plt.ylabel('3/x')
plt.legend(['y=3/x'])
# plt.text(x轴坐标，y轴坐标，添加的文本)
plt.text(12,0.7,'y=3/x')
plt.show()
```

## rc参数改变绘图样式

```python
data
```

```python
plt.figure(figsize=(8,5))
# 设置坐标轴颜色
plt.rcParams['axes.edgecolor']='m'
# 去除去除坐标轴
plt.rcParams['axes.spines.right']=False
plt.rcParams['axes.spines.top']=False
# 设置背景颜色
plt.rcParams['axes.facecolor']='c'
plt.plot(data,data,c='b',ls='--')
plt.title('y=x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 日历图

```python
import datetime
import random
from pyecharts.charts import Calendar
```

```python
begin=datetime.date(2023,1,1)
end=datetime.date(2023,12,31)
data=[
    [str(begin+datetime.timedelta(days=i)),random.randint(10000,20000)]
    for i in range((end-begin).days)
]
calendar=(
    Calendar(init_opts=opts.InitOpts(width="1200px",height="200px",
                                    theme=ThemeType.ESSOS))
    # calendar_opts=opts.CalendarOpts(range_=2023) 设置日历范围
    # time 表示系列名称
    # data 绘图数据，数据以键值对的形式存在
    .add('time',data,calendar_opts=opts.CalendarOpts(range_=2023))
    .set_global_opts(title_opts=opts.TitleOpts(
        {"title":"日历图"}),
                     visualmap_opts=opts.VisualMapOpts(
                         # is_piecewise=True 显示颜色区分
                         is_piecewise=True,min_=10000,max_=30000,
                         # pos_left="200px",pos_top="200px"设置图例位置
                         pos_left="200px",pos_top="200px"))
)
calendar.render("calendar.html")
calendar.render_notebook()
```

### 漏斗图

```python
from pyecharts.charts import Funnel
```

```python
data=[('浏览商品',50000),('放入购物车',35000),
      ('生成订单',25000),('支付订单',15000),('完成交易',8000)]
data
```

```python
funnel=(
    Funnel()
    .add('productions',data)
    .set_global_opts(title_opts=opts.TitleOpts(
        {"title":"漏斗图"}))
)
funnel.render('funnel.html')
funnel.render_notebook()
```
