```python
"""
author: Jackson Black
time: 2023-03-27 12-16-25 
"""

import pandas as pd
```

```python
# 1. 读取数据，将数据存成变量iris
data = pd.read_csv('iris.csv', header=None)
data.head()
```

```python
# 2. 创建数据框的列名称['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class’]
data.columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
data.head()
```

```python
# 3. 数据框中有缺失值吗？
# data.isnull().sum().sum()
# 第一个sum():判断每列空值并求出每一列空值的和
# 第二个sum():求出所有列的空值的和
data.isnull().sum().sum()
```

```python
# 4. 将列petal_length的第十到十九行设置为缺失值。
import numpy as np
data.loc[9:18,'petal_length']=np.nan
data.loc[9:18]
```

```python
# 5. 将petal_lengt缺失值全部替换为1.0。
data.fillna(1,inplace=True)
data.loc[9:18,:]
```

```python
# 6. 删除列class。
data.drop(columns='class',inplace=True)
data.head()
```

```python
# 7. 将数据框前三行设置为缺失值。
data.loc[0:2,:]=np.nan
data.head()
```

```python
# 8. 删除有缺失值的行。
data.dropna(inplace=True,how='any')
data
```

```python
# 9. 重新设置索引。
data.reset_index(inplace=True,drop=True)
data
```

```python
# 2. 查看前10行内容
data.head(10)
```

```python
# 3. 数据集中有多少个列(columns)？
data.shape[1]
```

```python
# 4. 打印出全部的列名称
data.columns
```

```python
# 6. 被下单数最多商品(item)是什么?
# 用于分组的字段不会存在于分组后的表中
data.groupby(by='item_name').agg({'quantity':'sum'}).idxmax()[0]
# data['item_name']
```

```python
# 7. 在item_name这一列中，一共有多少种商品被下单？
# unique()对数据去重
# nunique()在去重的基础上求和
data['item_name'].nunique()
```

```python
# 8. 一共有多少个商品被下单？
data['quantity'].sum()
```

```python
# 9. 将item_price转换为浮点数
# str  可以获取数据框中的所有字符窜数据
# astype  修改数据类型
data['item_price']=data['item_price'].str[1:].astype(float)
data['item_price']
```

```python
# 10. 在该数据集对应的时期内，收入(revenue)是多少？
data['income']=data['quantity']*data['item_price']
data['income'].sum()
```

```python
# 11. 在该数据集对应的时期内，一共有多少订单？
data['order_id'].nunique()
```

```python
# 12. 平均每一单(order)对应的总价是多少？
data['income'].sum()/data['order_id'].nunique()
```

```python
data.groupby(by='order_id').agg({'income':'sum'}).mean()[0]
```

## 练习三

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. 读取“appl_1980_2014.csv”数据并存为一个名叫apple的数据框。
data=pd.read_csv('appl_1980_2014.csv')
data.head()
```

```python
# 2. 查看每一列的数据类型。
data.dtypes
```

```python
# 3. 将Date这个列转换为datetime类型。
data['Date']=pd.to_datetime(data['Date'])
data.dtypes
```

```python
# 4. 将Date设置为索引。
data.set_index('Date',inplace=True)
data.head()
```

```python
# 5. 有重复的日期吗？
data.shape[0]==data.index.nunique()
```

```python
# 6. 将index设置为升序。
# ascending=True 表示进行升序排序
# inplace=True  表示对源数据进行修改
data.sort_index(inplace=True,ascending=True)
data.head()
```

```python
# 7. 找到每个月的最后一个交易日(businessday)
data.groupby(by=[data.index.year,data.index.month]
            ).agg({'Open':lambda x:x.index.day.max()})
```

```python
# 8. 数据集中最早的日期和最晚的日期相差多少天？
(data.index.max()-data.index.min()).days
```

```python
# 9. 在数据中一共有多少个月？
data.groupby(by=[data.index.year,data.index.month]
            ).agg({'Open':'count'}).shape[0]
```

```python
# 10. 按照时间顺序可视化Adj Close值。
plt.plot(data.index,data['Adj Close'])
```
